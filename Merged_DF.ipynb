# Merging multiple dataframes together in order to get a combined column

```python
import pandas as pd
```

```python
%store -r HnS_df
HnS_df.head()
```

```python
HnS_new_df = HnS_df.loc[:, ['Date','left_shoulder', 'right_shoulder', 'left_shoulder_diff', 'right_shoulder_diff', 'head_shoulders']]
HnS_new_df.head()
```

```python
%store -r Bollinger_df
Bollinger_df.head()
```

```python
merged_df = pd.merge(Bollinger_df, HnS_new_df, on='Date', how='inner')
```

```python
merged_df.head()
```

```python
%store -r kc_df
```

```python
kc_df = kc_df.loc[:, ['Date','kc_middle', 'kc_upper', 'kc_lower']]
kc_df.head()
```

```python
merged_df['Date'] = pd.to_datetime(merged_df['Date'])
merged_df_2 = pd.merge(kc_df, merged_df, on='Date', how='inner')
```

```python
merged_df_2.head()
```

```python
%store -r rsi_df
rsi_df.head()
```

```python
rsi_df['Date'] = pd.to_datetime(rsi_df['Date'])
merged_df_3 = pd.merge(rsi_df, merged_df_2, on='Date', how='inner')
```

```python
merged_df_3.head()
```

```python
%store -r ma_df
ma_df['Date'] = pd.to_datetime(ma_df['Date'])
merged_df_4 = pd.merge(ma_df, merged_df_3, on='Date', how='inner')
```

```python
merged_df_4.head()
```

```python
Max_Min_diff = merged_df_4.pop('Max_Min_diff')
merged_df_4.insert(len(merged_df_4.columns), 'Max_Min_diff', Max_Min_diff)

Open_Close_diff = merged_df_4.pop('Open_Close_diff')
merged_df_4.insert(len(merged_df_4.columns), 'Open_Close_diff', Open_Close_diff)

merged_df_4.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

def get_recommendation(predictions):
    return ["sell" if p == 0 else "buy" for p in predictions]


# define the features and label
features = ['MA', 'RSI', 'kc_middle', 'kc_upper', 'kc_lower', 'BB_upperband', 'BB_lowerband', 'left_shoulder', 'right_shoulder', 'left_shoulder_diff', 'right_shoulder_diff', 'head_shoulders', 'Max_Min_diff']
label = 'Buy or Sell'

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(merged_df_4[features], merged_df_4[label], test_size=0.2, random_state=42)

# create the random forest classifier and fit it to the training data
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)

# make predictions on the testing data
y_pred = rfc.predict(X_test)

# evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
